```python
def desc_cat(df):
    # Iterate through each column in the DataFrame
    for col in df.columns:
        # Check if the column is of object type (categorical)
        if df[col].dtype == 'object':
            print(f"--- Analysis for Column: {col} ---")
            print("Missing values:", df[col].isnull().sum())
            print("Total no of unique values:", df[col].nunique())
            print("Unique values:", df[col].unique())
            print("Mode:", df[col].mode()[0])
            print("Count of categories:\n", df[col].value_counts(), "\n")
```

```python
desc_cat(df)
```

```python
def desc_num(df):
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            print(f"                    --- Analysis for Column:{col} ---                    ")
            print("Missing values:", df[col].isnull().sum())
            print("Mean:", df[col].mean(), "    Median:", df[col].median(), "    Mode:", df[col].mode()[0])
            print("Range:", df[col].max() - df[col].min())
            q1 = df[col].quantile(0.25)
            q3 = df[col].quantile(0.75)
            iqr = q3 - q1
            
            print("IQR:", iqr)
            print("Standard Deviation:", df[col].std())
            lf = q1 - 1.5 * iqr
            uf = q3 + 1.5 * iqr
            condition_outlier = (df[col] < lf) | (df[col] > uf)
           
```

```python
desc_num(df)
```

# EDA

## Countplot for Gender

```python
sns.countplot(x='gender',data=df,palette="viridis")

#Rotate x-labels
plt.xticks(rotation=50);
```

## Customer Distribution by Country

```python
val_counts = df['country'].value_counts()
labels = val_counts.index
# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(val_counts, labels=labels, autopct='%1.2f%%', startangle=180, colors=plt.cm.Paired.colors)
plt.title('Customer Distribution by Country')
plt.show()
```

## Age Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True, bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## Boxplot of Tenure

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['tenure'], color='blue')
plt.title('Boxplot of Tenure')
plt.xlabel('Tenure')
plt.show()
```

## Relation between churn and gender

```python
fig, ax= plt.subplots(figsize= (10,8))
sns.countplot(x= 'gender', hue= 'churn', data= df)
plt.show()
```

## Countplot for Credit Card

```python
sns.countplot(x='credit_card',data=df)
plt.title('Countplot of Credit Card')
plt.ylabel('Count')
plt.xlabel('Credit Card')
plt.show()
```

## Distribution of Churn 

```python
sns.countplot(x='churn',data=df)
plt.title('Distribution of Churn')
plt.ylabel('Count')
plt.xlabel('Churn')
plt.show()
```

## Product Usage by Gender

```python
fig, ax= plt.subplots(figsize= (10,8))
sns.countplot(x= 'products_number', hue= 'churn', data= df)
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(12, 8))
corr_matrix = df[['credit_score', 'age', 'tenure', 'balance', 'products_number', 'estimated_salary']].corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
